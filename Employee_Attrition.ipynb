# Frequency Distribution 

```python
attrition_dist = round(df.Attrition.value_counts(normalize=True) * 100,2)
print('\nAttrition Distribution(%):')
attrition_dist
```

```python
gender_dist = round(df.Gender.value_counts(normalize=True) * 100,2)
print('\nGender Distribution:')
gender_dist
```

```python
department_dist = round(df.Department.value_counts(normalize=True) * 100,2)
print('\nDepartment Distribution:')
department_dist
```

```python
business_travel_dist = round(df.BusinessTravel.value_counts(normalize=True)*100,2)
print('\nBusinessTravel')
business_travel_dist
```

```python
education_feild = round(df.EducationField.value_counts(normalize=True)*100,2)
print('\nEducationField')
education_feild
```

```python
marital_status = round(df.MaritalStatus.value_counts(normalize=True)*100,2)
print('\nMaritalStatus')
marital_status
```

```python
job_role = round(df.JobRole.value_counts(normalize=True)*100,2)
print('\nJobRole')
job_role
```


# visualizing the distribution 

```python
fig, axes = plt.subplots(3,3, figsize=(18,15))
axes[0,0].pie(attrition_dist, labels=attrition_dist.index, autopct='%1.1f%%')
axes[0,0].set_title('Attrition Distribution');

axes[0,1].pie(gender_dist, labels=gender_dist.index, autopct='%1.1f%%')
axes[0,1].set_title('Gender Distribution');

axes[0,2].pie(department_dist, labels=department_dist.index, autopct='%1.1f%%')
axes[0,2].set_title('Department Distribution');

axes[1,0].pie(business_travel_dist, labels=business_travel_dist.index, autopct='%1.1f%%')
axes[1,0].set_title('Business Travel');

axes[1,1].pie(education_feild, labels=education_feild.index, autopct='%1.1f%%')
axes[1,1].set_title('Education Distribution');

axes[1,2].pie(marital_status, labels=marital_status.index, autopct='%1.1f%%')
axes[1,2].set_title('Marital Status');

axes[2,1].pie(job_role, labels=job_role.index, autopct='%1.1f%%')
axes[2,1].set_title('Job Role');

axes[2,0].axis('off');
axes[2,2].axis('off');
     
plt.tight_layout()
```

```python
df['count'] = 1
department_count = df.groupby(['Department','Attrition']).agg({'count':'sum'})
department_count
```

```python
#Dropping the count column 
```

```python
df.drop(columns='count', inplace=True)
```

# Attrition Percent in all Departments

```python
dept_att_count = df.groupby('Department')['Attrition'].value_counts(normalize=True) * 100
dept_att_count = dept_att_count.rename('Percentage').reset_index()

dept_att_pivot = dept_att_count.pivot(index='Department', columns='Attrition', values='Percentage').fillna(0)
dept_att_pivot = dept_att_pivot.reset_index()
print(dept_att_pivot)
```

# Attrition by Employee Job 

```python
df.JobRole.value_counts()
```

```python
jobrole = df.groupby('JobRole')['Attrition'].value_counts().unstack().fillna(0)
jobrole = jobrole.sort_values(by='Yes', ascending=False)
print(jobrole)
```

```python
plt.figure(figsize=(12,9))
plt.title('Attrition by Job Role')
sns.countplot(data=df,
             y='JobRole',
             hue='Attrition');
```
